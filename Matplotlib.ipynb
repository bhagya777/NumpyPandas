**86. Scatter Plot and Bar Plot**

*Making Figures Using Numpy Arrays*

```python
import numpy as np
```

```python
x=np.linspace(0,10,100)
x
```

```python
fig,ax=plt.subplots()
ax.plot(x,x**2);   #default line plot
plt.show()
```

```python
#Scatterplot
fig,ax=plt.subplots()
ax.scatter(x,np.exp(x))
plt.show()
```

```python
fig,ax=plt.subplots()
ax.scatter(x,np.sin(x))
plt.show()
```

*Making plots using dictionary*

```python
butter_prices={"a_butter":10,"p_butter":7,"c_butter":3}
fig,ax=plt.subplots()
ax.scatter(x=butter_prices.keys(),y=butter_prices.values())
ax.bar(butter_prices.keys(),height=butter_prices.values())
plt.show()
```

**87. Histograms And Subplots**

```python
fig,ax=plt.subplots()
ax.barh(list(butter_prices.keys()),list(butter_prices.values()))
plt.show()
```

```python
x=np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(x)
plt.show()
```

```python
#Subplots 1st way
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(butter_prices.keys(),butter_prices.values());
ax4.hist(np.random.randn(1000));
plt.show()
```

```python
#Subplots 2nd way
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(butter_prices.keys(),butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
plt.show()
```
