**Different ways of Plotting**

**1st**

```python
fig=plt.figure()
ax=fig.add_subplot()
plt.show()
```

**2nd**

```python
fig=plt.figure()
ax=fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

**3rd**

```python
fig,ax=plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
fig,ax=plt.subplots()
ax.plot(x,[10,40,160,250,340,520,700])
plt.show()
type(fig),type(ax)
```
